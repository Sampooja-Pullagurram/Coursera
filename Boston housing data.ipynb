# Is there a significant difference in median value of houses bounded by the Charles river or not? (T-test for independent samples)

Null hypothesis: There is no difference in average Median value of houses bounded & unbounded by the Charles river

Alternate hypothesis: There is a difference in average Median value of houses bounded & unbounded by the Charles river

```python
df.CHAS.value_counts()
```

```python
a=df[df.CHAS == 0]['MEDV']
a
```

```python
b=df[df.CHAS == 1]['MEDV']
b
```

```python
print(a.mean(), b.mean())
```

```python
scipy.stats.ttest_ind(a,b)
```

Since p-value less than alpha value of 0.05, we will reject null hypothesis. Therefore there is a statistical significance difference in average Median value of houses bounded & unbounded by the Charles river

# Is there a difference in Median values of houses (MEDV) for each proportion of owner occupied units built prior to 1940 (AGE)? (ANOVA)

Null hypothesis: There is no difference in average Median value of houses (MEDV) for each proportion of owner occupied units built prior to 1940 (AGE)

Alternate hypothesis: Atleast one of the average Median value of houses (MEDV) for each proportion of owner occupied units built prior to 1940 (AGE) differs

```python
df
```

```python
low=df[df.Age_group == '35 years and younger']['MEDV']
mid=df[df.Age_group == 'between 35 and 70 years']['MEDV']
high=df[df.Age_group == '70 years and older']['MEDV']
```

```python
f_statistic, p_value = scipy.stats.f_oneway(low, mid, high)
print('f_statistics={0}, p_value{1}'.format(f_statistic, p_value))
```

Since p_value is less than alpha value of 0.05, we will reject null hypothesis. Therefore atleast one of the average Median value of houses (MEDV) for each proportion of owner occupied units built prior to 1940 (AGE) differs

# Can we conclude that there is no relationship between Nitric oxide concentrations and proportion of non-retail business acres per town? (Pearson Correlation)

Null hypothesis: There is no correlation between NOX and INDUS

Alternate hypothesis: There is a correlation between NOX and INDUS

```python
pearson, p_value = scipy.stats.pearsonr(df.NOX, df.INDUS)
print('Pearson coefficient={0}, p_value={1}'.format(pearson, p_value))
```

Since p_value is less than alpha value of 0.05, we will reject null hypothesis.Therefore there exists a relationship between NOX and INDUS
